## Prepare Excel-file 'menu_items'

```python
import pandas as pd
import numpy as np
import os
```

```python
dataPath = r'C:\Users\gdawe2\OneDrive - GiGA\Mijn Documenten\Open Universiteit\Machine Learning\Opdracht\Data'
```

```python
data = pd.read_csv(os.path.join(dataPath,'fixations_selections_menu.csv'))
data
```

```python
def get_and_save_items():
    for ii in range(1,31):
        temp = list(data[(data['image']=='menu'+str(ii)+'.csv') & (data['Category']=='Menu_Item')]['Name'].unique())
        temp2 = list(data[(data['image']=='menu'+str(ii)+'.csv') & (data['selection_cat']=='Menu_Item')]['selection'].unique())
        items= list(set(temp+temp2))
        menu = ['menu'+str(ii)] * len(items)
        df = pd.DataFrame({'menu' : menu , 'items' : items})        
        if ii == 1: 
            df2 = df.copy()
        else:
            df2 = pd.concat([df2,df])
    df2 = df2.reset_index(drop=False)
    df2['index'] += 1
    df2.to_excel('menu_items0'+'.xlsx',index=False)
            
get_and_save_items()
```
